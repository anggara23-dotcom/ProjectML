```python
import pandas as pd
from pathlib import Path
import numpy as np
import os

# Corrected file path
xlsx_path = Path("/content/warehouse_supermart.xlsx")
out_dir = Path("/mnt/data/etl_output")
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
xls = pd.ExcelFile(xlsx_path)
dfs = {sheet: pd.read_excel(xls, sheet_name=sheet) for sheet in xls.sheet_names}
```

```python
def clean_strings(df):
    for c in df.select_dtypes(include=['object']).columns:
        df[c] = df[c].astype(str).str.strip().replace({'nan': None})
    return df

def normalize_gender(val):
    if pd.isna(val):
        return None
    s = str(val).strip().lower()
    if s in ('m', 'male', 'man', 'l', 'lk', 'laki', 'laki-laki'):
        return 'Male'
    if s in ('f', 'female', 'woman', 'perempuan'):
        return 'Female'
    return None  # unknown / invalid

def safe_to_numeric(s):
    try:
        if pd.isna(s): return None
        # remove common thousand separators and currency
        t = str(s).replace(',', '').replace(' ', '').replace('Rp', '').replace('$','')
        return float(t)
    except:
        return None
```

```python
# Clean dimension: dim_customer
if 'dim_customer' in dfs:
    df_cust = dfs['dim_customer'].copy()
    df_cust = clean_strings(df_cust)
    # Normalize gender to Male/Female/None
    if 'gender' in df_cust.columns:
        df_cust['gender'] = df_cust['gender'].apply(normalize_gender)
    # Parse registration_date if exists
    if 'registration_date' in df_cust.columns:
        df_cust['registration_date'] = pd.to_datetime(df_cust['registration_date'], errors='coerce')
    # ensure id is int
    if 'customer_id' in df_cust.columns:
        df_cust['customer_id'] = pd.to_numeric(df_cust['customer_id'], errors='coerce').astype('Int64')
    df_cust.to_csv(out_dir / "dim_customer_clean.csv", index=False)
else:
    df_cust = pd.DataFrame()
```

```python
if 'dim_produk' in dfs:
    df_prod = dfs['dim_produk'].copy()
    df_prod = clean_strings(df_prod)
    # price numeric
    if 'price' in df_prod.columns:
        df_prod['price'] = df_prod['price'].apply(safe_to_numeric)
    if 'product_id' in df_prod.columns:
        df_prod['product_id'] = pd.to_numeric(df_prod['product_id'], errors='coerce').astype('Int64')
    df_prod.to_csv(out_dir / "dim_produk_clean.csv", index=False)
else:
    df_prod = pd.DataFrame()
```

```python
if 'dim_store' in dfs:
    df_store = dfs['dim_store'].copy()
    df_store = clean_strings(df_store)
    if 'open_date' in df_store.columns:
        df_store['open_date'] = pd.to_datetime(df_store['open_date'], errors='coerce')
    if 'store_id' in df_store.columns:
        df_store['store_id'] = pd.to_numeric(df_store['store_id'], errors='coerce').astype('Int64')
    df_store.to_csv(out_dir / "dim_store_clean.csv", index=False)
else:
    df_store = pd.DataFrame()
```

```python
if 'dim_time' in dfs:
    df_time = dfs['dim_time'].copy()
    # Ensure date column parsed
    # Try to find a full date column
    if 'full_date' in df_time.columns:
        df_time['full_date'] = pd.to_datetime(df_time['full_date'], errors='coerce')
    if 'date_id' in df_time.columns:
        # ensure integer
        df_time['date_id'] = pd.to_numeric(df_time['date_id'], errors='coerce').astype('Int64')
    df_time.to_csv(out_dir / "dim_time_clean.csv", index=False)
else:
    df_time = pd.DataFrame()
```

```python
if 'fact_inventory' in dfs:
    df_inv = dfs['fact_inventory'].copy()
    df_inv = clean_strings(df_inv)
    # numeric conversions (try many column name variants)
    for col in df_inv.columns:
        if any(k in col.lower() for k in ['stock', 'min_', 'max_', 'ending_', 'out']):
            df_inv[col] = pd.to_numeric(df_inv[col], errors='coerce').astype('Int64', copy=False)
    if 'inventory_id' in df_inv.columns:
        df_inv['inventory_id'] = pd.to_numeric(df_inv['inventory_id'], errors='coerce').astype('Int64')
    df_inv.to_csv(out_dir / "fact_inventory_clean.csv", index=False)
else:
    df_inv = pd.DataFrame()
```

```python
# Clean fact_sales
if 'fact_sales' in dfs:
    df_sales = dfs['fact_sales'].copy()
    df_sales = clean_strings(df_sales)
    # numeric
    for col in ['quantity','unit_price','total_amount']:
        if col in df_sales.columns:
            df_sales[col] = pd.to_numeric(df_sales[col], errors='coerce')
    # ensure ids numeric
    for idcol in ['sales_id','date_id','product_id','customer_id','store_id']:
        if idcol in df_sales.columns:
            df_sales[idcol] = pd.to_numeric(df_sales[idcol], errors='coerce').astype('Int64')
    df_sales.to_csv(out_dir / "fact_sales_clean.csv", index=False)
else:
    df_sales = pd.DataFrame()
```

```python
# Referential integrity checks for fact_sales
fk_issues = []
if not df_sales.empty:
    # check product_id exists
    if not df_prod.empty and 'product_id' in df_prod.columns:
        prod_ids = set(df_prod['product_id'].dropna().astype(int).tolist())
        missing_prod = df_sales[~df_sales['product_id'].isna() & (~df_sales['product_id'].isin(prod_ids))]
        if not missing_prod.empty:
            fk_issues.append(("product_id", missing_prod.shape[0], missing_prod[['sales_id','product_id']].head(10)))
            missing_prod.to_csv(out_dir / "fact_sales_missing_product_fk.csv", index=False)
    # check customer_id exists
    if not df_cust.empty and 'customer_id' in df_cust.columns:
        cust_ids = set(df_cust['customer_id'].dropna().astype(int).tolist())
        missing_cust = df_sales[~df_sales['customer_id'].isna() & (~df_sales['customer_id'].isin(cust_ids))]
        if not missing_cust.empty:
            fk_issues.append(("customer_id", missing_cust.shape[0], missing_cust[['sales_id','customer_id']].head(10)))
            missing_cust.to_csv(out_dir / "fact_sales_missing_customer_fk.csv", index=False)
    # check store_id exists
    if not df_store.empty and 'store_id' in df_store.columns:
        store_ids = set(df_store['store_id'].dropna().astype(int).tolist())
        missing_store = df_sales[~df_sales['store_id'].isna() & (~df_sales['store_id'].isin(store_ids))]
        if not missing_store.empty:
            fk_issues.append(("store_id", missing_store.shape[0], missing_store[['sales_id','store_id']].head(10)))
            missing_store.to_csv(out_dir / "fact_sales_missing_store_fk.csv", index=False)
```

```python
report_lines = []
report_lines.append(f"dim_customer: {len(df_cust)} rows" if not df_cust.empty else "dim_customer: missing")
report_lines.append(f"dim_produk: {len(df_prod)} rows" if not df_prod.empty else "dim_produk: missing")
report_lines.append(f"dim_store: {len(df_store)} rows" if not df_store.empty else "dim_store: missing")
report_lines.append(f"dim_time: {len(df_time)} rows" if not df_time.empty else "dim_time: missing")
report_lines.append(f"fact_inventory: {len(df_inv)} rows" if not df_inv.empty else "fact_inventory: missing")
report_lines.append(f"fact_sales: {len(df_sales)} rows" if not df_sales.empty else "fact_sales: missing")
if fk_issues:
    report_lines.append("FK issues detected:")
    for col, cnt, sample in fk_issues:
        report_lines.append(f" - {col}: {cnt} missing (sample saved to etl_output)")
else:
    report_lines.append("No FK issues detected for fact_sales against dims.")

with open(out_dir / "etl_report.txt", "w") as f:
    f.write("\n".join(report_lines))
```

```python
# Create SQL DDL file (basic) for MySQL import using phpMyAdmin
ddl = f"""
-- DDL for db_penjualan_pipa (generated)
CREATE DATABASE IF NOT EXISTS db_penjualan_pipa;
USE db_penjualan_pipa;

CREATE TABLE dim_customer (
  customer_id INT PRIMARY KEY,
  customer_name VARCHAR(200),
  gender ENUM('Male','Female') DEFAULT NULL,
  age_group VARCHAR(50),
  city VARCHAR(100),
  membership_type VARCHAR(50),
  registration_date DATE
);

CREATE TABLE dim_produk (
  product_id INT PRIMARY KEY,
  product_name VARCHAR(200),
  category VARCHAR(100),
  subcategory VARCHAR(100),
  brand VARCHAR(100),
  price DECIMAL(12,2),
  supplier_name VARCHAR(200)
);

CREATE TABLE dim_store (
  store_id INT PRIMARY KEY,
  store_name VARCHAR(200),
  city VARCHAR(100),
  region VARCHAR(100),
  store_size VARCHAR(50),
  manager_name VARCHAR(100),
  open_date DATE
);

CREATE TABLE dim_time (
  date_id INT PRIMARY KEY,
  full_date DATE,
  day INT,
  month INT,
  year INT,
  quarter INT,
  day_name VARCHAR(20),
  month_name VARCHAR(20),
  is_weekend TINYINT(1)
);

CREATE TABLE fact_inventory (
  inventory_id INT PRIMARY KEY,
  date_id INT,
  product_id INT,
  store_id INT,
  stock_in INT,
  stock_out INT,
  ending_stock INT,
  min_stock_level INT,
  max_stock_level INT
);

CREATE TABLE fact_sales (
  sales_id INT PRIMARY KEY,
  date_id INT,
  product_id INT,
  customer_id INT,
  store_id INT,
  quantity INT,
  unit_price DECIMAL(12,2),
  total_amount DECIMAL(14,2),
  payment_method VARCHAR(50)
);

-- Add FK constraints (optional)
ALTER TABLE fact_sales ADD FOREIGN KEY (product_id) REFERENCES dim_produk(product_id);
ALTER TABLE fact_sales ADD FOREIGN KEY (customer_id) REFERENCES dim_customer(customer_id);
ALTER TABLE fact_sales ADD FOREIGN KEY (store_id) REFERENCES dim_store(store_id);
ALTER TABLE fact_sales ADD FOREIGN KEY (date_id) REFERENCES dim_time(date_id);
"""

with open(out_dir / "ddl_db_penjualan_pipa.sql", "w") as f:
    f.write(ddl)

print("ETL cleaning finished. Cleaned CSVs and report saved to:", out_dir)
print("\nFiles created:")
for p in sorted(out_dir.glob("*")):
    print(" -", p.name)
```
